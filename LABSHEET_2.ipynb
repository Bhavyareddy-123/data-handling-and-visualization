```python
df=pd.read_csv("/content/drive/MyDrive/data (1).csv")
print(df.head)
print(df.to_markdown())
```

```python
# Mode Method
x=df["Calories"].mode()[0]
df["Calories"].fillna(x,inplace=True)
print(df.to_markdown())
```

```python
#Mean method
df=pd.read_csv("/content/drive/MyDrive/data (1).csv")
print(df.head)

```

```python
x=df["Calories"].mean()
df["Calories"].fillna(x,inplace=True)
print(df.to_markdown())
```

```python
#Median method
df=pd.read_csv("/content/drive/MyDrive/data (1).csv")
print(df.head)

```
